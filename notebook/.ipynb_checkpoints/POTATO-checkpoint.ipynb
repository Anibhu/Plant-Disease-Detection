```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetV2B3
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, TensorBoard, CSVLogger
import cv2
# from tensorflow.keras import layers, models
import keras
from keras import layers
from keras import models
from sklearn.metrics import classification_report, confusion_matrix
import os
import datetime
# from tensorflow.keras import layers, models, regularizers
from keras import regularizers
from keras.callbacks import EarlyStopping
from sklearn.utils.class_weight import compute_class_weight
import pathlib
```

```python
# Constants
BATCH_SIZE = 32
IMAGE_SIZE = 300  # EfficientNetV2B3 prefers slightly larger images
CHANNELS = 3
EPOCHS = 30
FINE_TUNE_EPOCHS = 15
BASE_DIR = r"C:\Users\ANISH BHUIN\OneDrive\Desktop\jbooks\Potato"
MODEL_PATH = "potato_leaf_disease_model_effnet.keras"
```

```python
# Load pre-split datasets
def load_pre_split_datasets(base_dir):
    train_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(base_dir, "Train"),
        seed=123,
        image_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE
    )
    
    val_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(base_dir, "Valid"),
        seed=123,
        image_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE
    )
    
    test_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(base_dir, "Test"),
        seed=123,
        image_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE
    )
    
    return train_ds, val_ds, test_ds

train_ds, val_ds, test_ds = load_pre_split_datasets(BASE_DIR)
class_names = train_ds.class_names
n_classes = len(class_names)

print(f"Class names: {class_names}")
print(f"Train samples: {len(train_ds)*BATCH_SIZE}")
print(f"Validation samples: {len(val_ds)*BATCH_SIZE}")
print(f"Test samples: {len(test_ds)*BATCH_SIZE}")
```

```python
# Optimize dataset performance
train_ds = train_ds.cache().shuffle(1000).prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.cache().prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.cache().prefetch(tf.data.AUTOTUNE)
```

```python
# Data augmentation
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2),
    layers.RandomTranslation(height_factor=0.1, width_factor=0.1),
    layers.RandomBrightness(0.2),
    layers.GaussianNoise(0.1),
])
```

```python
# Apply augmentation to training set
train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y),
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)
```

```python
# Hyperparameter tuning with Keras Tuner
def build_model(hp):
    base_model = EfficientNetV2B3(
        input_shape=(IMAGE_SIZE, IMAGE_SIZE, CHANNELS),
        include_top=False,
        weights='imagenet',
        pooling='avg'
    )
    base_model.trainable = False
    
    inputs = keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, CHANNELS))
    x = data_augmentation(inputs)
    x = tf.keras.applications.efficientnet_v2.preprocess_input(x)
    x = base_model(x)
    
    # Tunable dropout rate
    x = layers.Dropout(hp.Float('dropout_1', 0.3, 0.7, step=0.1))(x)
    
    # Tunable dense units
    x = layers.Dense(
        hp.Int('dense_units_1', 256, 1024, step=128),
        activation='relu',
        kernel_regularizer=regularizers.l2(hp.Float('l2_1', 1e-5, 1e-2, sampling='log'))
    )(x)
    x = layers.BatchNormalization()(x)
    
    x = layers.Dropout(hp.Float('dropout_2', 0.2, 0.5, step=0.1))(x)
    
    outputs = layers.Dense(n_classes, activation='softmax')(x)
    
    model = keras.Model(inputs, outputs)
    
    # Tunable learning rate
    lr = hp.Float('learning_rate', 1e-5, 1e-3, sampling='log')
    optimizer = keras.optimizers.Adam(learning_rate=lr)
    
    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

```python
# Compute class weights
def extract_labels(dataset):
    all_labels = []
    for _, labels in dataset.unbatch():
        all_labels.append(labels.numpy())
    return np.array(all_labels)

train_labels = extract_labels(train_ds)
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_labels),
    y=train_labels
)
class_weight_dict = dict(enumerate(class_weights))
print("Class weights:", class_weight_dict)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, TensorBoard
import datetime

def get_callbacks():
    return [
        EarlyStopping(
            monitor='val_loss',
            patience=6,
            restore_best_weights=True,
            verbose=1
        ),
        ModelCheckpoint(
            'best_model_tuning.keras',  # ✅ use .keras, not .h5
            monitor='val_accuracy',
            save_best_only=True,
            mode='max'
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=1e-7,
            verbose=1
        ),
        TensorBoard(
            log_dir="logs/tuning/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
        )
    ]
```

```python
import keras_tuner as kt
# Hyperparameter search
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=20,
    factor=3,
    directory='tuning',
    project_name='potato_disease'
)
print("Total trials: ", tuner.oracle.max_trials)
print("Starting hyperparameter search...")
# tuner.search(
#     train_ds,
#     validation_data=val_ds,
#     epochs=20,
#     callbacks=get_callbacks(),
#     class_weight=class_weight_dict
# )
tuner.search(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    class_weight=dict(enumerate(class_weights)),  # Ensures it's pickleable
    callbacks=get_callbacks()
)

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
Best hyperparameters:
- dropout_1: {best_hps.get('dropout_1')}
- dense_units_1: {best_hps.get('dense_units_1')}
- l2_1: {best_hps.get('l2_1')}
- dropout_2: {best_hps.get('dropout_2')}
- learning_rate: {best_hps.get('learning_rate')}
""")

# Build and train final model with best hyperparameters
model = tuner.hypermodel.build(best_hps)
```

```python
# Train initial model
print("Training initial model...")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=get_callbacks(),
    class_weight=class_weight_dict
)
```

```python
# Fine-tuning
print("\nFine-tuning model...")
# Step 1: Get the base model
base_model = model.layers[2]  # 'efficientnetv2-b3' is at index 2

# Step 2: Unfreeze the base model
base_model.trainable = True

# Step 3: Freeze the first 150 layers (to avoid overfitting or catastrophic forgetting)
for layer in base_model.layers[:150]:
    layer.trainable = False
```

```python
# Recompile with lower learning rate
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=best_hps.get('learning_rate')/10),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train fine-tuned model
fine_tune_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=FINE_TUNE_EPOCHS,
    callbacks=get_callbacks(),
    class_weight=class_weight_dict
)
```

```python
# Evaluation
print("\nEvaluating model...")
test_loss, test_acc = model.evaluate(test_ds)
print(f"\nTest accuracy: {test_acc:.4f}")
print(f"Test loss: {test_loss:.4f}")
```

```python
def display_gradcam(img, heatmap, alpha=0.4):
    # Rescale heatmap to a range 0-255
    heatmap = np.uint8(255 * heatmap)
    
    # Use jet colormap to colorize heatmap
    jet = plt.cm.get_cmap("jet")
    
    # Use RGB values of the colormap
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]
    
    # Create an image with RGB colorized heatmap
    jet_heatmap = keras.preprocessing.image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = keras.preprocessing.image.img_to_array(jet_heatmap)
    
    # Superimpose the heatmap on original image
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = keras.preprocessing.image.array_to_img(superimposed_img)
    
    # Display Grad CAM
    plt.figure(figsize=(10, 10))
    plt.imshow(superimposed_img)
    plt.axis('off')
    plt.show()

# Get sample images for Grad-CAM
sample_images = []
sample_labels = []
for images, labels in test_ds.take(1):
    sample_images = images[:3]
    sample_labels = labels[:3]
    break

# Preprocess images
preprocessed_images = []
for img in sample_images:
    img_array = tf.keras.applications.efficientnet_v2.preprocess_input(
        tf.expand_dims(img, axis=0)
    )
    preprocessed_images.append(img_array)

# Last conv layer name in EfficientNetV2B3
last_conv_layer_name = "top_conv"

# Generate Grad-CAM heatmaps
for i, (img, img_array) in enumerate(zip(sample_images, preprocessed_images)):
    # Generate class activation heatmap
    preds = model.predict(img_array)
    pred_index = np.argmax(preds[0])
    
    # ✅ Pass the base model for Grad-CAM
    heatmap = make_gradcam_heatmap(
        img_array,
        model.get_layer("efficientnetv2-b3"),
        last_conv_layer_name,
        pred_index
    )
    
    # Display original image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img.numpy().astype("uint8"))
    plt.title(f"Original\nTrue: {class_names[sample_labels[i]]}\nPred: {class_names[pred_index]}")
    plt.axis('off')
    
    # Display heatmap
    plt.subplot(1, 2, 2)
    plt.imshow(heatmap, cmap='jet')
    plt.title("Attention Heatmap")
    plt.axis('off')
    plt.show()
```

```python
# Classification report and confusion matrix
print("\nGenerating classification report...")
y_true = []
y_pred = []

for images, labels in test_ds.unbatch():
    y_true.append(labels.numpy())
    predictions = model.predict(tf.expand_dims(images, axis=0))
    y_pred.append(np.argmax(predictions, axis=1)[0])

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))
```

```python
# Confusion matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
